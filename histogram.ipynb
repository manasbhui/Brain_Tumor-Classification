```python
import cv2
import matplotlib.pyplot as plt

# Load the images
glioma_image = cv2.imread('C:/Users/Manas/Desktop/OBJECT DETECTION/BT/Test/substracted image/glioma_substracted/G_2_VF_.jpg', cv2.IMREAD_GRAYSCALE)
meningioma_image = cv2.imread('C:/Users/Manas/Desktop/OBJECT DETECTION/BT/Test/substracted image/meningioma_substracted/M_2_VF_.jpg', cv2.IMREAD_GRAYSCALE)
pituitary_image = cv2.imread('C:/Users/Manas/Desktop/OBJECT DETECTION/BT/Test/substracted image/pituitary_substracted/P_2_VF_.jpg', cv2.IMREAD_GRAYSCALE)

# Display the images
plt.figure(figsize=(12, 4))
plt.subplot(131), plt.imshow(glioma_image, cmap='gray'), plt.title('Glioma')
plt.subplot(132), plt.imshow(meningioma_image, cmap='gray'), plt.title('Meningioma')
plt.subplot(133), plt.imshow(pituitary_image, cmap='gray'), plt.title('Pituitary')
plt.show()
```

```python
# Generate histograms
hist_glioma = cv2.calcHist([glioma_image], [0], None, [256], [0, 256])
hist_meningioma = cv2.calcHist([meningioma_image], [0], None, [256], [0, 256])
hist_pituitary = cv2.calcHist([pituitary_image], [0], None, [256], [0, 256])

# Plot histograms
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.plot(hist_glioma)
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(132)
plt.plot(hist_meningioma)
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(133)
plt.plot(hist_pituitary)
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.tight_layout()  # Ensure plots don't overlap
plt.show()

```
